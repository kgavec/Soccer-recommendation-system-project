## <center> Juntar DB's Transfermrkt & SOFIFA 

#### <i> FUNCION PARA CALCULAR LA SIMILITUD DE DOS STRING : VALOR DE 0.0 A 1.0

```python
def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```python
def Generar_ids(dftransf,dfsofifa,id_start):
    #-----Crear dataframe temporal que servira de Base -> DATA TRANSFERMARKT
    temp_df=pd.DataFrame()
    #Insertar columna de ID
    temp_df.insert(0, 'ID', range(1+id_start, len(dftransf)+1+id_start))
    #Insertar columna de nombres de jugadores
    temp_df.insert(1,'name',dftransf['name'].values)
    #insertarfecha
    temp_df.insert(2,'birth_date',dftransf['birthdate'].values)
    
    #----Crear dataframe temporal que servira de Base -> DATA SOFIFA
    tempo=pd.DataFrame()
    #Crear dataframe temporal para la data de SOFIFA
    tempo.insert(0, 'full_name', dfsofifa['full_name'])
    tempo.insert(1, 'nombre_init', dfsofifa['nombre_init'])
    #AÑADIR una columna concatenada de nombre completo con abreviatura 
    tempo['concat'] = tempo['full_name'] +'-'+tempo['nombre_init']
    #insertar  fecha
    tempo.insert(3, 'Fec_nac', dfsofifa['Fec_nac'])
    
    
    #Crear dataframes que se usaran para comparar similitudes
    DB_TRANSFERMARKT=temp_df[['name','birth_date']]
    #sofifa
    DB_SOFIFA=tempo[['concat','Fec_nac']]


    #Comparar similitudes de nombres de TRANSFERMARKT con SOFIFA 
    #Crear listas
    option_a_1=[]
    option_a_2=[]
    option_b_1=[]
    option_b_2=[]
    option_c=[]
    #Por cada nombre de jugador de transfermrkt
    for i,transf in enumerate(DB_TRANSFERMARKT['name']):
        #Iniciar en Falso para las listas
        full_1=False
        full_2=False
        f_1=False
        f_2=False
        f_c=False
        #Por cada nombre de jugador de sofifa
        for x,sofifa in enumerate(DB_SOFIFA['concat']):
            #similitud en fecha_nacimiento
            date_val=similar(DB_SOFIFA.iloc[x][1],DB_TRANSFERMARKT.iloc[i][1])
            #similitud en nombre
            name_val1=similar(transf,sofifa.split("-")[0])
            name_val2=similar(transf,sofifa.split("-")[1])#shortname
            
            #Comparar si el indicador de similitud es mayor a 0.65
            if similar(transf,sofifa.split("-")[0])>=0.65:
                #
                if full_1==False:
                    if date_val>0.6:
                        option_a_1.append(sofifa.split("-")[0])
                        full_1=True
                elif full_2==False:
                    if date_val>0.6:
                        option_a_2.append(sofifa.split("-")[0])
                        full_2=True
            if similar(transf,sofifa.split("-")[1])>=0.65:
                if f_1==False:
                    if date_val>0.6:
                        option_b_1.append(sofifa.split("-")[1])
                        f_1=True
                elif f_2==False:
                    if date_val>0.6:
                        option_b_2.append(sofifa.split("-")[1])
                        f_2=True
            if similar(transf,sofifa)>=0.65:
                if f_c==False:
                    option_c.append(sofifa)
                    f_c==True
        #Evaluar si la lista obtuvo una coincidencia de nombre; de lo contrario llenar ese espacio con ""             
        if len(option_a_1)<i+1:
            #print("na")
            option_a_1.append("")
        if len(option_a_2)<i+1:
            #print("na")
            option_a_2.append("")
        if len(option_b_1)<i+1:
            #print("na")
            option_b_1.append("")
        if len(option_b_2)<i+1:
            #print("na")
            option_b_2.append("")
        if len(option_c)<i+1:
            #print("na")
            option_c.append("")
            
    #Insertar datos similares al df usado como base 
    temp_df.insert(2,'option_1a',option_a_1)
    temp_df.insert(3,'option_2a',option_a_2)
    temp_df.insert(4,'option_1b',option_b_1)
    temp_df.insert(5,'option_2b',option_b_2)
    try:
        temp_df.insert(6,'option_c',option_c)
    except:
        print(len(option_c))
    #Eliminar columnas team y birth_Date
    temp_df= temp_df.drop("birth_date", axis=1)
    
    return temp_df
```

#### Funcion evaluar el string más aproximado al nombre real

```python
def evaluate(df_w_options):
    #numero de filas
    num_rows=df_w_options.shape[0]
    #evaluadores definitivos
    op1_def=[]
    op2_def=[]

    #Agregar columna de verificacion de nombre 
    for row in range(0,num_rows):
        #Extraer datos a evaluar de una misma fila
        name=df_w_options.iloc[row][1]
        opa1=df_w_options.iloc[row][2]
        opa2=df_w_options.iloc[row][3]
        opb1=df_w_options.iloc[row][4]
        opb2=df_w_options.iloc[row][5]
        
        #Hacer evaluacion
        if similar(name,opa1)>similar(name,opa2):
            op1_def.append(opa1)
        elif similar(name,opa1)<similar(name,opa2):
            op1_def.append(opa2)
        elif similar(name,opa1)==similar(name,opa2):
            op1_def.append(opa1)
        else:
            op1_def.append("")

        if similar(name,opb1)>similar(name,opb2):
            op2_def.append(opb1)
        elif similar(name,opb1)<similar(name,opb2):
            op2_def.append(opb2)
        elif similar(name,opb1)==similar(name,opb2):
            op2_def.append(opb1)
        else:
            op2_def.append("")



    #Insertar resultados
    df_w_options.insert(2,'res_1',op1_def)
    df_w_options.insert(3,'res_2',op2_def)

    #Eliminar las otras columnas
    df_w_options= df_w_options.drop("option_1a", axis=1)
    df_w_options= df_w_options.drop("option_2a", axis=1)
    df_w_options= df_w_options.drop("option_1b", axis=1)
    df_w_options= df_w_options.drop("option_2b", axis=1)
    df_w_options= df_w_options.drop("option_c", axis=1)
    
    return df_w_options
```

```python
def final_eval(df_result_1_2):
    num_rows=df_result_1_2.shape[0]
    final=[]

    #Elegir con que columna se hará el merge
    for row in range(0,num_rows):
        name=df_result_1_2.iloc[row][1]
        op1_def=df_result_1_2.iloc[row][2]
        op2_def=df_result_1_2.iloc[row][3]
        if similar(name,op1_def)>similar(name,op2_def):
            final.append("1")
        elif similar(name,op1_def)<similar(name,op2_def):
            final.append("2")
        elif similar(name,op1_def)==similar(name,op2_def):
            final.append("2")
        else:
            final.append("")

    df_result_1_2.insert(4,'final',final)
    return df_result_1_2
```

```python
def split_df(df_last):
    num_rows=df_last.shape[0]
    
    #Generar DF1
    df_1=pd.DataFrame()
    p1={'ID': '1','name': 'Ederson','full_nam': 'Ederson Mike'}#un valor random de un jugador
    #Asignar estructura a Dataframe
    df_1 = df_1.assign(**p1) #asignar cabeceras de acuerdo al diccionario armado
    
    #Generar DF2
    df_2=pd.DataFrame()
    p2={'ID': '1','name': 'Ederson','short_nam': 'E. Mike'}#un valor random de un jugador
    #Asignar estructura a Dataframe
    df_2 = df_2.assign(**p2) #asignar cabeceras de acuerdo al diccionario armado
    
    #Elegir con que columna se hará el merge
    for row in range(0,num_rows):
        id_df=df_last.iloc[row][0]
        name=df_last.iloc[row][1]
        op1_def=df_last.iloc[row][2]
        op2_def=df_last.iloc[row][3]
        final=df_last.iloc[row][4]
        
        if final=="1":
            data={'ID': id_df,'name': name,'full_nam': op1_def}
            df_1 = df_1.append(data, ignore_index=True)
        elif final=='2':
            data={'ID': id_df,'name': name,'short_nam': op2_def}
            df_2 = df_2.append(data, ignore_index=True)
        
    return df_1, df_2
```

```python
def full_dataset(res1,res2,df_Transf,df_SoFifa):
    #Unir splits de dataset 
    #resultado1 con #sofifa
    full_data=pd.merge(res1, df_SoFifa, how='left',left_on="full_nam",right_on="full_name")
    full_data= full_data.drop("full_nam", axis=1)
    #resultado1 con #transfermrkt
    full_data=pd.merge(full_data, df_Transf, how='left',on="name")

    #Unir splits de dataset 
    #resultado2 con #sofifa
    full_data2=pd.merge(res2, df_SoFifa, how='left',left_on="short_nam",right_on="nombre_init")
    #eliminar col
    full_data2= full_data2.drop("short_nam", axis=1)
    #resultado2 con #transfermrkt
    full_data2=pd.merge(full_data2, df_Transf, how='left',on="name")
    
    #concatenar datasets
    final=pd.concat([full_data, full_data2], ignore_index=True)
    #ordenar
    final=final.sort_values(['ID'],ascending=True)
    #restear index
    final.index = pd.RangeIndex(len(final.index))
    
    return final
```

```python
def generar_fullDB(df_transfer,df_SOFIFA,ID_init):
    #Generar tabla inicial con ID, nombre de jugador (transfermarkt), y evaluaciones de opciones de coincidencia de nombre de SOFIFA
    Init=Generar_ids(df_transfer,df_SOFIFA,ID_init)
    #Evaluar opciones generedas nuevamente
    fase_one=evaluate(Init)
    #Asignar 1 o 2 de acuerdo a que columna se usará como key para hacer el merge 
    fase_two=final_eval(fase_one)
    #Hacer split de los dataframe para poder cruzarlos con los db's 
    split1,split2=split_df(fase_two)
    #Hacer los cruces respectivos para generar el dataset completo 
    FULL_DATA=full_dataset(split1,split2,df_transfer,df_SOFIFA)
    
    return FULL_DATA
```
